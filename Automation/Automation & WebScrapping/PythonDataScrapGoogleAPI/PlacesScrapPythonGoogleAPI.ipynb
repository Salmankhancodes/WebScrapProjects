```python
import pandas as pd
import requests
import time
import json
DataPlaces=[]
```

```python
keywords=['hospital']
radius='2000'
api_key='Enter your Api_key'
coordinates=['28.362253, 79.419456']
```

```python
for cor in coordinates:
    for kw in keywords:
        url='https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+cor+'&radius='+str(radius)+'&keyword='+str(kw)+'&key='+str(api_key)
        while True:
            print(url)
            res=requests.get(url)
            js=json.loads(res.text)
            results=js['results']
            for result in results:
                name=result['name']
                place_id=result['place_id']
                types=result['types']
                rating=result['rating']
                vicinity=result['vicinity']
                data = [name, place_id, types, rating, vicinity]
                DataPlaces.append(data)
                time.sleep(5)
            if 'next_page_token' not in js:
                break
            else:
                next_page_token = js['next_page_token']
                url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?key='+str(api_key)+'&pagetoken='+str(next_page_token)
                 
labels=['Hospital Name','ID','types','Ratings','Address']            
df=pd.DataFrame(DataPlaces,columns=labels)
```
