```python
from selenium import webdriver
```

```python
url="https://webscraper.io/test-sites/e-commerce/static"
driver=webdriver.Chrome()
driver.get(url)
```

```python
driver.find_element_by_xpath('//*[@id="side-menu"]/li[2]/a').click()
driver.find_element_by_xpath('//*[@id="side-menu"]/li[2]/ul/li[1]/a').click()
```

```python
condition=True
allinks=[]
while (condition):
    products=driver.find_elements_by_class_name('thumbnail')
    for item in products:
        b=item.find_elements_by_tag_name('h4')[-1]
        c=b.find_element_by_tag_name('a')
        d=c.get_property('href')
        allinks.append(d)
    try:
        nextP=driver.find_elements_by_class_name('page-link')[-1]
        if nextP.get_attribute('aria-label')=='Next »':
            nextP.click()
        else:
            condition=False
    except:
        condition=False
```

```python
from tqdm import tqdm
dataa=[]
for i in tqdm(allinks):
    driver.get(i)
    Prod=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[2]').text
    Pri=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[1]').text
    Riv=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[2]/p').text
    Des=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/p').text
    data={
    'Product':Prod,
    'Price':Pri,
    'Rivew':Riv,
    'Description':Des,
    'link':i
    }
    dataa.append(data)
```

```python
import pandas as pd
finalResult=pd.DataFrame(dataa)
finalResult.to_csv('OutputExample.csv')
finalResult
```
